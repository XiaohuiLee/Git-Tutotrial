# Scenario 1: INITIALIZE A REPOSITORY FROM EXSTING COEE

````python
git init
````

This command would generate a .git directory in the folder where we execute this command. Output may come as bellow:

````python
Initialized empty Git repository in E:/OneDrive/_GitRepo/local_repo/.git/
````

.git is a directory that contains all we need to track our codes, so if you don't want to track the project any more, you just type following command in the git bash

````python
rm -rf .git
````

### ADD FILE TO STAGING AREA

````python
# add all files to the staging area
git add -A

# add specific file to staging area
git add filename

git status
````

See the output as bellow:

````python
 $ git add -A
warning: LF will be replaced by CRLF in .gitignore.
The file will have its original line endings in your working directory.


$ git status
On branch master

Initial commit

Changes to be committed:
  (use "git rm --cached <file>..." to unstage)

        new file:   .gitignore
        new file:   NLP.py
````

### REMOVE FILES FROM STAGING AREA

````python
git reset
git reset NLP.py

git status
# next cell shows the output from the terminal
````
